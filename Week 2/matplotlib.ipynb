## Line Plot
Kapan digunakan ?
- Untuk melihat perubahan nilai waktu ke waktu -> aplikasi treding
- Time series data

```python
x = [1, 2, 3, 4, 5]
y = [10, 35, 20 ,70, 40]

plt.Figure(figsize=(8, 4))
# plt.plot()
plt.plot(x, y)

# Detail element
plt.title("Contoh Line Plot")
plt.xlabel("Sumbu X")
plt.ylabel("Sumbu y")
plt.grid(True)
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 35, 20 ,70, 40]
y2 = [15, 20, 50, 25, 15]

plt.Figure(figsize=(8, 4))
# plt.plot()
plt.plot(x, y, marker="o", color="purple", linestyle="-", label="MTK")
plt.plot(x, y2, marker="o", color="orange", linestyle="--", label="Fisika")

# Detail element
plt.title("Contoh Line Plot")
plt.xlabel("Sumbu X")
plt.ylabel("Sumbu y")
plt.grid(True)
plt.legend()
```

## Bar Plot
Digunakan untuk:
- Membandingkan nilai antar kategori.

```python
kategori = ["Python", "Java", "PHP", "Ruby", "R", "C++"]
jumlah_pengguna = [100, 10, 15, 120, 80, 70]

plt.figure(figsize=(8, 5))
# plt.bar()
plt.bar(kategori, jumlah_pengguna, label= "Jumlah Pengguna")
plt.xlabel("Bahasa Pemrograman")
plt.ylabel("Jumlah pengguna")
plt.title("Jumlah Pengguna Bahasa Pemrograman")
plt.grid(axis="y", alpha=0.5, linestyle="--")
```

```python
kategori = ["Batch 1", "Batch 2", "Batch 3"]
lakilaki = [40, 60, 55]
perempuan = [ 15, 30, 50]

x = np.arange(len(kategori))
barwidth = 0.35

plt.figure(figsize=(8, 5))
# plt.bar()
plt.bar(x - barwidth/2, lakilaki, width=barwidth, label="Laki-laki")
plt.bar(x + barwidth/2, perempuan, width=barwidth, label="Perempuan")

plt.xlabel("Bahasa Pemrograman")
plt.ylabel("Jumlah pengguna")
plt.title("Jumlah Pengguna Bahasa Pemrograman")
plt.grid(axis="y", alpha=0.5, linestyle="--")
plt.xticks(x, kategori)
plt.legend()
plt.tight_layout()
```

## Scatter Plot
Digunakan untuk:
1. Menampilkan hubungan antara 2 variabel

```python
umur = [18, 21, 25, 30, 35, 40, 45, 50]
penghasilan = [1.5, 2.2, 2.8, 4.0, 4.2, 4.5, 4.7, 5.0]

plt.figure(figsize=(8, 5))
# plt.scatter()
plt.scatter(umur, penghasilan)
plt.title("Hubungan antara umur dan penghasilan")
plt.xlabel("Umur (tahun)")
plt.ylabel("Pengahasilan dalam (juga)")
plt.grid(True)
plt.tight_layout()
```

```python
fisika_A = [75, 80, 85, 90, 78]
matematika_A = [70, 85, 82, 90, 99]

fisika_B = [60, 55, 61, 20,90]
matematika_B = [65, 62, 70, 80, 11]

plt.figure(figsize=(8, 5))
# plt.scatter()
plt.scatter(fisika_A, matematika_A, label="Nilai Kelas A")
plt.scatter(fisika_B, matematika_B, label="Nilai Kelas B")
plt.title("Hubungan antara fisika dan matematika")
plt.xlabel("Nilai Fisika")
plt.ylabel("Matematika")
plt.grid(True, alpha=0.5)
plt.tight_layout()
plt.legend(loc="lower left")
```
