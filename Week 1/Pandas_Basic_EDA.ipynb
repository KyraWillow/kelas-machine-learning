## Advance Data Preprocessing

Pada sub ini kita akan mencoba untuk menangani sebuah `outlier` menggunakan rumus matematika.

```python
df_numbers = df.select_dtypes(include=("int64", "float64")).columns.to_list()
```

```python
# Membuat sebuah fungsi untuk menangai outlier
def remove_outlier (df, columns, x):
    df_clean = df.copy()
    for cols in columns:
        Q1 = df_clean[cols].quantile(0.25)
        Q3 = df_clean[cols].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - x * IQR
        upper_bound = Q3 + x * IQR
        df_clean = df_clean[(df_clean[cols] >= lower_bound) & (df_clean[cols] <= upper_bound)]
    return df_clean
```

```python
df_no_outliers = remove_outlier(df, df_numbers, 1.5)
df_no_outliers
```

### Normalisasi

- Metode `MinMaxScaler()` = Mengubah skala `min` menjadi 0 & `max` menjadi 1 **(Hanya berfungsi jika datanya sudah normal/tidak ada outlier)**
- Metode `StandardScaler()` = Mengubah skala `mean` mendekati 0 & `std` mendekati 1 **(Hanya berfungsi jika datanya sudah normal/tidak ada outlier)**
- Metode `RobustScaler()` = Mengubah skala `min` menjadi 0 & `max` menjadi 1 **(Berfungsi juga apabila terdapat outlier di dalam data/tahan terhadap outlier)**

#### MinMaxScaler()

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
minmax = MinMaxScaler()
```

```python
df_min_max_scaled = df_no_outliers.copy()
```

```python
df_min_max_scaled[df_numbers] = minmax.fit_transform(df_no_outliers[df_numbers])
```

```python
df_min_max_scaled.describe()
```
